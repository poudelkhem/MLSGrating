```python
import numpy as np
import matplotlib.pyplot as plt
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 25}

```

```python
def cleanDataFunction(filename1):
    lines = filename1.readlines()[8:]
    lineArray=[]
    angle=[]
    reflection=[]
    for l in lines:
        
        words=l.strip().split()
        lineArray.append(words)
    lineArray=np.array(lineArray)
    for w in lineArray:
        angle.append(float(w[0]))
        #print(angle)
        reflection.append(float(w[1]))
    return angle,reflection
    
        
```

```python
MLS1_330000=open("MLS1_330000.txt","r")# open the file name "ReflectionMultilayer_625.txt"
angleMLS1_330000,reflectionMLS1_330000=cleanDataFunction(MLS1_330000)# call the cleanDataFunction
MLS1_330005=open("MLS1_330005.txt","r")# open the file name "ReflectionMultilayer_625.txt"
angleMLS1_330005,reflectionMLS1_330005=cleanDataFunction(MLS1_330005)# call the cleanDataFunction

Periodic1_33=open("Periodic1_33.txt","r")# open the file name "ReflectionMultilayer_625.txt"
anglePeriodic1_33,reflectionPeriodic1_33=cleanDataFunction(Periodic1_33)# call the cleanDataFunction
```

```python
plt.plot(anglePeriodic1_33, reflectionPeriodic1_33,'r',linewidth=3,label="Periodic1_33")
plt.plot(angleMLS1_330000, reflectionMLS1_330000,'b--',linewidth=3,label="MLS1_33")
#plt.plot(angleMLS1_330005, reflectionMLS1_330005,'r',linewidth=3,label="MLS1_330")



plt.legend(bbox_to_anchor=(1.0, 1), loc=2, borderaxespad=0.)
#plt.legend(loc=0)
plt.ylabel("Reflectivity",fontsize=26)
plt.xlabel("Angle (Deg)",fontsize=26)
#plt.gca().set_xlim(left=6.10, right=6.15)
plt.rc('font', **font)
plt.grid('on')
plt.show()
```
