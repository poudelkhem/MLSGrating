```python
import numpy as np
import matplotlib.pyplot as plt
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 25}

```

```python
def cleanDataFunction(filename1):
    lines = filename1.readlines()[8:]
    lineArray=[]
    angle=[]
    reflection=[]
    for l in lines:
        
        words=l.strip().split()
        lineArray.append(words)
    lineArray=np.array(lineArray)
    for w in lineArray:
        angle.append(float(w[0]))
        #print(angle)
        reflection.append(float(w[1]))
    return angle,reflection
    
        
```

```python

MLS1_3280=open("MLS1_3280.txt","r")# open the file name "ReflectionMultilayer_625.txt"
angleMLS1_3280,reflectionMLS1_3280=cleanDataFunction(MLS1_3280)# call the cleanDataFunction

MLS1_3285=open("MLS1_3285.txt","r")# open the file name "ReflectionMultilayer_625.txt"
angleMLS1_3285,reflectionMLS1_3285=cleanDataFunction(MLS1_3285)# call the cleanDataFunction

MLS1_3290=open("MLS1_3290.txt","r")# open the file name "ReflectionMultilayer_625.txt"
angleMLS1_3290,reflectionMLS1_3290=cleanDataFunction(MLS1_3290)# call the cleanDataFunction

MLS1_3295=open("MLS1_3295.txt","r")# open the file name "ReflectionMultilayer_625.txt"
angleMLS1_3295,reflectionMLS1_3295=cleanDataFunction(MLS1_3295)# call the cleanDataFunction

MLS1_330000=open("MLS1_33.txt","r")# open the file name "ReflectionMultilayer_625.txt"
angleMLS1_330000,reflectionMLS1_330000=cleanDataFunction(MLS1_330000)# call the cleanDataFunction
#plt.plot(angleMLS1_3275, reflectionMLS1_3275,'r',linewidth=3,label="R.I = 1.3275")


 
maxpos13280 = reflectionMLS1_3280.index(max(reflectionMLS1_3280))  
xc13280=angleMLS1_330000[maxpos13280]

maxpos13285 = reflectionMLS1_3285.index(max(reflectionMLS1_3285))  
xc13285=angleMLS1_330000[maxpos13285]

maxpos13290 = reflectionMLS1_3290.index(max(reflectionMLS1_3290))  
xc13290=angleMLS1_330000[maxpos13290]

maxpos13295 = reflectionMLS1_3295.index(max(reflectionMLS1_3295))  
xc13295=angleMLS1_330000[maxpos13295]

maxpos13300 = reflectionMLS1_330000.index(max(reflectionMLS1_330000))  
xc13300=angleMLS1_330000[maxpos13300]
xc13300
```
